```python
from diffusers import DiffusionPipeline
import torch
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)
```

```python
pipe = DiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5"
)

pipe = pipe.to("cpu")
```

```python
prompt = "A futuristic smart city at sunset, ultra realistic"

image = pipe(
    prompt,
    num_inference_steps=20
).images[0]

image.save("smart_city.png")

image
```

```python
prompt = "A cyberpunk robot walking in Tokyo, cinematic lighting"
image = pipe(
    prompt,
    num_inference_steps=20
).images[0]

image.save("smart_city.png")

image
```
